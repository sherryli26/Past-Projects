```python
import pandas as pd
import numpy as np
```

```python

stopwords = pd.read_csv("Data/stop_words.csv")
stopwords = list(stopwords['word'])
```

```python
# writing a function to open file
def op(name):
    with open(f"Data/{name}-khashoggi.txt","r") as file:
        news=file.read() #question: why can't the file.read() be saved in a new variable "news" if I do not return it?
    return news

alj= op("aljazeera")
bbc= op("bbc")
bre= op("breitbart")
cnn=op("cnn")
fox=op("fox")


```

```python
#writing a function to get rid of punctuations and stopwords and transform the string to a list of words
def transform(name):
    name=name.lower()
    for i in [",",".","'","-","(",")","\"","“","—"]:
        name=name.replace(i,"")
    name= name.split()
    k=[]#generating a new list to store the list without stopwords

#    for i in stopwords:
#        if i in name: name =name.remove(i) #why does this generate error?
    
    for i in name:
        if i not in stopwords:k.append(i)
    return k
    

    
```

```python
alj=transform(alj)
bbc=transform(bbc)
bre=transform(bre)
cnn=transform(cnn)
fox=transform(fox)
```

```python
#writing a function to count the frequncy of words
def count(transformed):
    """
    This function counts the frequncy of words.
    """
    d=dict()
    for word in transformed:
        if word in d:
            d[word][0]+=1
        else:
            d[word]=[1]
    
    return d
```

```python
#writing a function to put the counted numbers in the same dataframe
def bind(results):
    """
    This function takes in the previously generated list of words and use the count() function 
    we defined above to count the word frequncy, and put the result of five medias in one dataframe.
    """
    DTM=pd.DataFrame()
    for r in results:
        dtf=pd.DataFrame(count(r))
        DTM=DTM.append(dtf,ignore_index=True,sort=True) #row bind
    DTM.fillna(0,inplace=True)
    return DTM

final= bind([alj,bbc,bre,cnn,fox])
```

```python
def cosine(a,b):
    cos= np.dot(a,b)/np.sqrt(np.dot(a,a))/np.sqrt(np.dot(b,b))
    return cos
```

```python
# using cosine to calculate the similarity between medias
cosdf=pd.DataFrame(index=["aljazeera","bbc","breitbart","cnn","fox"],columns=["aljazeera","bbc","breitbart","cnn","fox"])
for i in range(5):
    for a in range(5):
        cosdf.iloc[i,a]=cosine(final.iloc[i].values,final.iloc[a].values)

cosdf
```

BBC,aljazeera and fox reported the news in a similar way. Breitbart	and CNN reported in a less similar way compared with other medias.
